```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np
```

The state of the reservoir is updated at every time step $dt$ according to
$$ 
\mathbf{r}(t + dt) = f[\mathbf{W}_{res} \cdot \mathbf{r}(t) + 
\mathbf{W}_{IR} \cdot \mathbf{u}(t)]
$$

```python
class Reservoir(layers.Layer):
    
    def __init__(self, units, spectral_radius, alpha=1, degree=3, **kwargs):
        kwargs["trainable"] = False
        super(Reservoir, self).__init__(**kwargs)
        self.__alpha = abs(alpha)
        self.__units = units
        
        # generating adjoint matrix w_res
        row_idx = np.tile(np.arange(units, dtype=np.int64), [degree])
        col_idx = np.random.permutation(row_idx)
        val = np.random.random(units * degree)
        w_res = np.zeros((units, units), dtype=np.float32)
        w_res[row_idx, col_idx] = val  
        rho = np.abs(np.linalg.eigvals(w_res)).max()
        w_res = spectral_radius * w_res / rho
        self.w_res = tf.sparse.from_dense(w_res)
        self.cold_start()
            
    def cold_start(self):
        self.res = tf.zeros((1, self.__units), dtype=tf.dtypes.float32)
        
    def build(self, input_shape):
        self.w_ir = tf.random.uniform((input_shape[-1], self.__units),
                                      minval=-self.__alpha, maxval=self.__alpha,
                                      dtype=tf.float32)
        
    def call(self, inputs):
        self.res = tf.transpose(
            tf.sparse.sparse_dense_matmul(self.w_res, self.res,
                                          adjoint_a=True, adjoint_b=True))
        self.res += tf.matmul(inputs, self.w_ir)
        out = keras.activations.tanh(self.res)
        return tf.concat([out, out * out], -1)
```

```python
class ChaosNet(object):
    
    def __init__(self, *args, **k2args):
        self.reservoir = Reservoir(*args, **k2args)
        self.is_trained = False
        self.__gamma = 1e-4
        
    def train(self, train_x, train_y):
        train_x = tf.expand_dims(train_x, 1)
        self.reservoir.cold_start()
        out = tf.concat([self.reservoir(x) for x in train_x], 0)
        self.w_ro = tf.linalg.inv((tf.transpose(out) @ out + self.__gamma * tf.eye(out.shape[-1]))) @ tf.transpose(out) @ train_y
        self.is_trained = True
        print(f"training finished using {train_y.shape[0]} sample points")
    

cha = ChaosNet(4992, 0.1)
cha.train(train_x, train_y)
```

```python
def predict(self, initial_x, step=1000):
    if self.is_trained:
        self.reservoir.cold_start()
        out = [initial_x]
        for t in range(step):
            initial_x = self.reservoir(initial_x) @ self.w_ro
            out.append(initial_x)
        return tf.concat(out, 0)

pred = predict(cha, tf.expand_dims(validate_y[0], 0))
```

## Kuramoto-Sivashinsky Equation

The KSE is a 1D nonlinear PDE given by
$$
\frac{\partial u}{\partial t}=
-u \frac{\partial u}{\partial x} - \frac{\partial^2 u}{\partial x^2}
-\frac{\partial^4 u}{\partial x^4}
$$

```python
kse_data = loadmat("KSE_code/KS_data_right.mat")["udata"]
train_x = tf.constant(kse_data[:, 2001:72010].T, dtype=tf.float32)
train_y = tf.constant(kse_data[:, 2002:72011].T, dtype=tf.float32)
validate_x = kse_data[:, 72011:73210]
validate_y = tf.constant(kse_data[:, 72012:73211].T, dtype=tf.float32)
```

```python
%matplotlib inline
plt.figure(figsize=(8, 2.5))
plt.imshow(tf.transpose(validate_y), origin="lower", 
           extent=[0, validate_y.shape[0]*0.25*0.05, 0, 22], 
           aspect="auto", cmap="jet")
plt.colorbar()
plt.ylabel("$x$")
plt.xlabel("$\Lambda_m t$")
plt.show()
```

```python
# ridge regression
gamma = 1e-4
res = Reservoir(3000, 0.1)
res_x = res(train_x)
weight = tf.linalg.inv((tf.transpose(res_x) @ res_x + gamma * tf.eye(res_x.shape[-1]))) @ tf.transpose(res_x) @ (train_y - train_x)
```

```python
pred_y = res_x @ weight + train_x
plt.figure(figsize=(8, 2.5))
plt.imshow(tf.transpose(pred_y-train_y)[:, :2000], origin="lower", 
           extent=[0, validate_y.shape[-1]*0.25*0.05, 0, 22], 
           aspect="auto", cmap="jet")
plt.colorbar()
plt.ylabel("$x$")
plt.xlabel("$\Lambda_m t$")
plt.show()
```
